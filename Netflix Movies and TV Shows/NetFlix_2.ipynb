```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('netflix_titles.csv')
data.head()
```

```python
data.info()
```

```python
data.drop('director',axis =1, inplace = True)
```

```python
data.columns
```

```python
data.dropna(inplace = True)
data.info() # here still the index is taking from 1 to 8806
```

```python
data.dropna(inplace = True)
```

```python
data.reset_index(drop = True, inplace = True)
```

```python
data.isnull().sum()
data
```

```python
data.country.values
```

```python
unique_country_name = []
for country in data.country.values:
    # print(country)
    country_name = country.split(',')
    for countries in country_name:
        countries = countries.strip()
        if countries not in unique_country_name:
            unique_country_name.append(countries)
```

```python
print(unique_country_name)
print(len(unique_country_name))
```

```python
data.shape
```

```python
empty_country_df = np.zeros((data.shape[0], len(unique_country_name)))
```

```python
empty_country_df.shape
```

```python
#new dataframe

country_df = pd.DataFrame(empty_country_df, columns = unique_country_name)
```

```python
country_df
```

```python
 data.country = data.country.apply(lambda x: x.split(','))
```

```python
data.country = data.country.apply(lambda x: [i.strip() for i in x])
print(data.country)
print(data.country.values)
```

```python
for idx, values in enumerate(data.country.values):
    for country in values:
        country_df.loc[idx, country] = 1
```

```python
country_df
```

```python
country_data = pd.concat([data, country_df], axis = 1)
country_data.drop('country', axis = 1, inplace = True)
```

```python
top10 = country_data.iloc[:, 10:].sum(axis = 0).sort_values(ascending = False)[:10].index.tolist()
```

```python
d = dict()
for coun in top10:
    d[coun] = country_data.loc[country_data[coun] == 1, 'type'].value_counts(normalize = True)
```
