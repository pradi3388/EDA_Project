### How many movies rleased btn 2000 to 2005 which is more than 90 minutes length belongs to comedies genre

```python
only_comedies = []
for listed_in in data.listed_in.values:
    # print(listed_in)
    comedyGenre = listed_in.split(',')
    for name in comedyGenre:
        name = name.strip()
        if name not in only_comedies:
            only_comedies.append(name)
```

```python
print(only_comedies)
print(len(only_comedies))
```

```python
data.shape
```

```python
empty_genre_df = np.zeros((data.shape[0], len(only_comedies)))
empty_genre_df.shape 
```

```python
genre_df = pd.DataFrame(empty_genre_df,columns = only_comedies)
genre_df
```

```python
data.listed_in = data.listed_in.apply(lambda x: x.split(','))
data.listed_in = data.listed_in.apply(lambda x: [i.strip() for i in x])
data.listed_in
```

```python
data.listed_in.values
```

```python
for idx, values in enumerate(data.listed_in.values):
    for listed_in in values:
        genre_df.loc[idx, listed_in] = 1
```

```python
genre_df
```

```python
genre_data = pd.concat([data,genre_df], axis =1)
genre_data.drop('listed_in', axis = 1, inplace = True)
genre_data
```

```python
genre_data1 = genre_data.loc[genre_data['Comedies'] == 1, ['type', 'release_year','duration']]
genre_data1 = genre_data1.loc[(genre_data1.release_year >= 2000) & 
                              (genre_data1.release_year <=2005), :]
```

```python
genre_data1.duration = genre_data1.duration.str.replace(r'[a-zA-Z ]', '', regex = True)
```

```python
genre_data1.duration = genre_data1.duration.apply(lambda x: int(x))
```

```python
genre_data1
newData = genre_data1.loc[(genre_data1.duration >= 90)]
newData
```

```python
newData.value_counts()
```

```python
newData.drop('type', axis = 1, inplace = True)
```

```python
newData.release_year.value_counts().plot(kind= 'bar')
```

```python
sns.barplot(x = newData.release_year.value_counts().index, 
            y = newData.release_year.value_counts().values,
            palette='viridis')
```

```python
import pandas as pd
import numpy as np
dummy = pd.DataFrame({
        'animals': ['cat', 'dog', 'cow', 'cat', 'cow'],
        'food':['fish,milk', 'bones,chicken,fish', 'grass,chicken', 'fish,milk', 'grass']
})
```

```python
dummy
```

```python
unique_vals = []
for i in dummy.food.values:
    unique_vals.extend(i.split(','))
```

```python
unique_vals = list(set(unique_vals))
print(unique_vals)
```

```python
dummy.shape[0]
```

```python
new_df_val = np.zeros((dummy.shape[0], len(unique_vals)))
```

```python
new_df = pd.DataFrame(new_df_val, columns = unique_vals)
```

```python
new_df
```

```python
for idx, i in enumerate(dummy.food.values):
    val = i.split(',')
    for j in val:
        new_df.loc[idx, j] = 1
```
