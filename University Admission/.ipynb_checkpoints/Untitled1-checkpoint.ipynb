**PEARSONS CORELATION**

```python
# Pearson corelation is applicable for neumarical values and not for categorical values and it varies from -1 to +1 and heat map is used to plot the matrix
# if there is a corelation between 2 coloumns will drop anyone coloumn
# While creating a machine learning model, if corelation between 2 features(coloumns) is really high then it significe that the features are
# dependent but we need all the features are independent so we can drop 1 of the 2 features
# hex plot is similar to scatter plot
```

```python
new_df = df[['GRE Score', 'TOEFL Score','CGPA', 'Chance of Admit']]
new_df
```

```python
new_df.corr()
```

```python
sns.heatmap(new_df.corr(), annot = True)
```

```python
sns.jointplot(x = 'CGPA', y = 'Chance of Admit', data = new_df, kind = 'hex')
```
